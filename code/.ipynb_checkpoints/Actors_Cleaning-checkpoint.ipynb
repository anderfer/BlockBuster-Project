```python
## Buscar nulos y duplicados

def check_suma_nulos(df: pd.DataFrame)-> None:   ## Suma de valores nulos por columna
    
    nan_cols = df.isna().sum()
    
    nan_cols = nan_cols[nan_cols>0]
    
    display(nan_cols)
    
check_suma_nulos(actors)
```
