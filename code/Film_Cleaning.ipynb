```python
def check_nulos(df: pd.DataFrame)-> None:  ## Buscar % de nulos
    
    nan_cols = df.isna().mean() * 100  # % de valores nulos
    
    nan_cols = nan_cols[nan_cols>0]
    
    display(f'N nan cols: {len(nan_cols)}')
    display(nan_cols)
    
check_nulos(film)
```

```python
## Cambiar el rating a texto para entenderlo mejor

mapeo = {'G': 'General Audience', 'PG': 'Parental Guidance Suggested', 'PG-13': 'Parents Strongly Cautioned', 'R': 'Restricted', 'NC-17': 'Adults Only'}
film['rating'] = film['rating'].replace(mapeo)
film
```

```python
film.special_features.unique()
```

```python
# Creo una columna por cada valor unico de la columna special_features
film['Deleted Scenes'] = 'No'
film['Behind the Scenes'] = 'No'
film['Trailers'] = 'No'
film['Commentaries'] = 'No'

# Si la columna special_features contiene la palabra de las nuevas columnas, se pone un Yes en la columna nueva.

for index, row in film.iterrows():
    if 'Deleted Scenes' in row['special_features']:
        film.at[index, 'Deleted Scenes'] = 'Yes'
    if 'Behind the Scenes' in row['special_features']:
        film.at[index, 'Behind the Scenes'] = 'Yes'
    if 'Trailers' in row['special_features']:
        film.at[index, 'Trailers'] = 'Yes'
    if 'Commentaries' in row['special_features']:
        film.at[index, 'Commentaries'] = 'Yes'
        
# Elimino la columna original
film = film.drop('special_features', axis=1)


```

```python
language_origin = pd.read_csv('C:/Users/ferna/Desktop/IRONHACK/BlockBuster-Project/data/language.csv')
language = language_origin.copy()

```

```python
## todas las peliculas tienen idioma ingles, por lo que prescindo de la tabla languages y lo sustituyo por una columna con el idioma

film = film.merge(language[['language_id', 'name']], on='language_id', how='left')
```

```python
## Pongo el nombre de la columna bien y elimino el language_id

film = film.rename(columns={'name': 'language'})
film = film.drop(columns=['language_id'])
```
