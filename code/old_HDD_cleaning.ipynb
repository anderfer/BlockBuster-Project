```python
def check_nulos(df: pd.DataFrame)-> None:  ## Buscar % de nulos
    
    nan_cols = df.isna().mean() * 100  # % de valores nulos
    
    nan_cols = nan_cols[nan_cols>0]
    
    display(f'N nan cols: {len(nan_cols)}')
    display(nan_cols)
    
check_nulos(old_HDD)
```

```python
## Como en esta tabla tengo nombres y apellidos de artistas voy a usarlo para ponerle el actor_id

actors_origin = pd.read_csv('C:/Users/ferna/Desktop/IRONHACK/BlockBuster-Project/data/actor.csv')
actors = actors_origin.copy()

old_HDD['actor_id'] = None

for index, row in old_HDD.iterrows():
    last_name = row['last_name']
    first_name = row['first_name']
    
    matching_actor = actors[(actors['last_name'] == last_name) & (actors['first_name'] == first_name)]
    
    if not matching_actor.empty:
        actor_id = matching_actor['actor_id'].values[0]
        old_HDD.at[index, 'actor_id'] = actor_id
        
        
old_HDD
```

```python
actors
```

```python
## veo que hay un actor duplicado por nombre y apellido, lo elimino.

actors = actors.drop_duplicates(subset=['last_name', 'first_name'], keep="first")
actors

```

```python
## Como en esta tabla tengo titulos voy a usarlo para ponerle el film_id

film_origin = pd.read_csv('C:/Users/ferna/Desktop/IRONHACK/BlockBuster-Project/data/film_curated.csv')
film = film_origin.copy()

old_HDD['film_id'] = None

for index, row in old_HDD.iterrows():
    title = row['title']
    
    matching_title = film[(film['title'] == title)]
    
    if not matching_title.empty:
        title_id = matching_title['film_id'].values[0]
        old_HDD.at[index, 'film_id'] = title_id

old_HDD
```

```python
## Filtro por film ID <= 223, la cantidad de peliculas en el inventario, para poder actualizar nombres que no tenga

old_HDD = old_HDD[old_HDD['film_id'] <= 223]
old_HDD.film_id.unique()
```
